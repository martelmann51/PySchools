```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete




```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate Total Number of Schools
TotalSchools = school_data_complete["school_name"].nunique()

TotalSchools
```

```python
#Calculate Total Number of Students of District
TotalStudents = int(school_data_complete["Student ID"].nunique())

TotalStudents
```

```python
#Calculate Total Budget of District
UniqueBudget = school_data_complete["budget"].unique()


TotalBudget = int(UniqueBudget.sum())

TotalBudget
```

```python
#Calculate Per Student Budget

PerStudentBudget = (int(TotalBudget/TotalStudents))


PerStudentBudget
```

```python
#Calculate Average Math Score

AverageMathScore = round(school_data_complete["math_score"].mean(),2)

AverageMathScore
```

```python
#Calculate Average Reading Score

AverageReadingScore = round(school_data_complete["reading_score"].mean(),2)

AverageReadingScore
```

```python
#Calculate Overall Passing Rate

OverallPassingRate = round((AverageMathScore + AverageReadingScore)/2,2)

OverallPassingRate
```

```python
PassingMath = school_data_complete.loc[school_data_complete["math_score"] >= 70]
StudentsPassingMath = PassingMath["student_name"].nunique()

StudentsPassingMath
```

```python
#Percentage of Students with Passing Math Scores

Percentage_Passing_Math = round((StudentsPassingMath/TotalStudents)*100,2)

Percentage_Passing_Math
```

```python
PassingReading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
StudentsPassingReading = PassingReading["student_name"].nunique()

StudentsPassingReading
```

```python
#Percentage of Students with Passing Reading Scores

Percentage_Passing_Reading = round((StudentsPassingReading/TotalStudents)*100,2)

Percentage_Passing_Reading
```

```python
#DistrictSummary

district_summary_df = pd.DataFrame({"Total Schools": [TotalSchools], 
                                    "Total Students": [TotalStudents], 
                                    "Total Budget": [TotalBudget],
                                    "Budget Per Student": [PerStudentBudget],
                                    "Average Math Score": [AverageMathScore],
                                    "Average Reading Score": [AverageReadingScore],
                                    "Overall Passing Rate": [OverallPassingRate],
                                    "Percentage of Students Passing Math": [Percentage_Passing_Math],
                                    "Percentage of Students Passing Reading": [Percentage_Passing_Reading]
                                   })
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("{:,.2f}".format)
district_summary_df["Budget Per Student"] = district_summary_df["Budget Per Student"].map("{:,.2f}".format)
district_summary_df["Percentage of Students Passing Math"] = district_summary_df["Percentage of Students Passing Math"].map("{:}%".format)
district_summary_df["Percentage of Students Passing Reading"] = district_summary_df["Percentage of Students Passing Reading"].map("{:}%".format)
district_summary_df
                                
district_summary_df
```

```python
#Per Student Budget


group_school_data = school_data_complete.groupby(["school_name"])
Total_Students = group_school_data["student_name"].count()
Average_Student_Math_Scores = round(group_school_data["math_score"].mean(),2)
Average_Student_Reading_Scores = round(group_school_data["reading_score"].mean(),2)
School_Types = school_data.set_index("school_name")["type"]
Total_School_Budget = round(school_data.set_index("school_name")["budget"],2)
Budget_Per_Student = round(Total_School_Budget/Total_Students,2)
Passing_Reading = student_data.loc[student_data["reading_score"] >= 70]
Group_Passing_Reading = Passing_Reading.groupby(["school_name"])
Count_Passing_Reading = Group_Passing_Reading["reading_score"].count()
Percent_Passing_Reading = round((Count_Passing_Reading/Total_Students*100),2)
Passing_Math = student_data.loc[student_data["math_score"] >=70]
Group_Passing_Math = Passing_Math.groupby(["school_name"])
Count_Passing_Math = Group_Passing_Math["math_score"].count()
Percent_Passing_Math = round((Count_Passing_Math/Total_Students*100),2)
Overall_Passing_Rate = round(((Percent_Passing_Reading + Percent_Passing_Math)/2),2)





```

```python
#School Summary

school_summary_df = pd.DataFrame({"School Type": School_Types,
                                  "Total Students": Total_Students,
                                  "Total School Budget": Total_School_Budget,
                                  "Budget Per Student": Budget_Per_Student,
                                    "Average Math Score": Average_Student_Math_Scores,
                                     "Average Reading Score": Average_Student_Reading_Scores,
                                     "Percent Passing Reading": Percent_Passing_Reading,
                                     "Percent Passing Math": Percent_Passing_Math,
                                     "Overall Passing Rate": Overall_Passing_Rate})

school_summary_df["Total Students"] = school_summary_df["Total Students"]
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("{:,.2f}".format)
school_summary_df["Budget Per Student"] = school_summary_df[("Budget Per Student")]
school_summary_df["Percent Passing Math"] = school_summary_df["Percent Passing Math"]
school_summary_df["Percent Passing Reading"] = school_summary_df["Percent Passing Reading"]
school_summary_df["Overall Passing Rate"] = school_summary_df["Overall Passing Rate"]


school_summary_df
```
